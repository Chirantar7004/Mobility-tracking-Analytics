```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('another2.csv', dtype='object')
```

```python
info_column = df['Info'].str.split(',', expand=True) # split whenever a comma found
df['Type'] = info_column[0]
df['SN'] = info_column[1]
df['FN'] = info_column[2]
df['Flags'] = info_column[3]
df['BI'] = info_column[4]
df['SSID'] = info_column[5]
df.drop(columns=['Info'], inplace=True)
df
```

```python
types = df.groupby('Type').Source.count()
types
probes = df.query('Type == "Probe Request"')
probes.count
probes
```

```python
df.Time = pd.to_numeric(df.Time, errors='coerce')
```

```python
labells=['0.5hr','1hr','1.5hr','2hr','2.5hr','3hr','3.5hr','4hr','4.5hr','5hr','5.30hr','6hr','6.5hr','7hr']
fd = df.groupby(pd.cut(df['Time'], bins=[0,1800,3600,5400,7200,9000,10800,12600,14000,16200,18000,19800,21600,23400,25200],labels=labells))
rq=fd.Time.count().to_dict()
rq
```

```python
names = list(rq.keys())
values = list(rq.values())

plt.bar(range(len(rq)),values,tick_label=names,width=0.3,color=['black', 'red', 'green', 'blue', 'cyan','yellow','orange'])
plt.savefig('bar.png')
plt.xticks(rotation=90)
plt.xlabel('Hours')
plt.ylabel('No. of Probe Requests')
plt.show()
```
